Stacked Bar chart

```python
plt.bar(df['PetalLengthCm'],df['SepalWidthCm'],label='Group 1',color='blue')
plt.bar(df['SepalLengthCm'],df['PetalWidthCm'],bottom=df['PetalLengthCm'],label='Group 2',color='pink')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Stacked Bar Chart')
plt.legend()
plt.show()
```

3D visualization

```python
x=df['PetalLengthCm']
y=df['PetalWidthCm']
z=df['SepalLengthCm']
fig=go.Figure(data=[go.Scatter3d(
    x=x,y=y,z=z,
    marker=dict(size=5,color='yellow',colorscale='Viridis'),
    line=dict(color='black',width=3)
)])
fig.update_layout(title='3D Scatter Plot')
fig.show()
```
